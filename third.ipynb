```python
#  Explain: Data Cleaning & Preprocessing 
# • Load the dataset and check for missing values. 
# • Fill missing Age values with median or mean. 
# • Handle missing Embarked values by filling with the most common port. 
# • Drop unnecessary columns like Name, Ticket, Cabin for visualization purposes.  

# Data Cleaning & Preprocessing means preparing raw data so it becomes suitable for analysis or visualization.
# It includes:

# Handling missing values

# Removing unnecessary columns

# Fixing data types

# Preparing data for modeling

# We will use pandas for this task.
import pandas as pd

# Load dataset (example: Titanic dataset CSV)
df = pd.read_csv("titanic.csv")

# View first 5 rows
print(df.head())

# Check missing values
print("\nMissing Values:\n")
print(df.isnull().sum())

# Fill Age with median (recommended)
df["Age"].fillna(df["Age"].median(), inplace=True)

# OR use mean
# df["Age"].fillna(df["Age"].mean(), inplace=True)

# Find most common value (mode)
most_common_port = df["Embarked"].mode()[0]

# Fill missing values
df["Embarked"].fillna(most_common_port, inplace=True)

df.drop(columns=["Name", "Ticket", "Cabin"], inplace=True)

print("\nMissing Values After Cleaning:\n")
print(df.isnull().sum())

```

```python
# Univariate Analysis 
# • Plot the distribution of Age using a histogram and a KDE plot. 
# • Plot the count of passengers by Sex using a bar chart. 
# • Show the distribution of Fare using a boxplot.

import pandas as pd
import matplotlib.pyplot as plt

# Sample Titanic-like dataset
data = {
    "Age": [22, 38, 26, 35, 28, 2, 54, 27, 14, 4, 58, 20, 39, 30, 25],
    "Sex": ["male", "female", "female", "female", "male", "male",
            "male", "male", "female", "male", "female",
            "female", "male", "male", "female"],
    "Fare": [7.25, 71.83, 7.92, 53.1, 8.05, 21.07, 51.86,
             11.13, 30.07, 16.7, 26.55, 13.0, 35.5, 8.46, 18.75]
}

df = pd.DataFrame(data)

# 1. Histogram of Age
plt.figure()
plt.hist(df["Age"], bins=8)
plt.title("Age Distribution (Histogram)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# 2. KDE Plot of Age
plt.figure()
df["Age"].plot(kind="kde")
plt.title("Age Distribution (KDE)")
plt.xlabel("Age")
plt.show()

# 3. Bar Chart of Passenger Count by Sex
plt.figure()
df["Sex"].value_counts().plot(kind="bar")
plt.title("Passenger Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()

# 4. Boxplot of Fare
plt.figure()
plt.boxplot(df["Fare"])
plt.title("Fare Distribution (Boxplot)")
plt.ylabel("Fare")
plt.show()
```

```python
# Bivariate Analysis 
# • Compare survival rate across Sex using a countplot. 
# • Compare Pclass vs Fare using a boxplot. 
# • Visualize Age vs Fare using a scatter plot, color-coded by Survived.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")

plt.figure()
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Sex")
plt.show()

plt.figure()
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Pclass vs Fare")
plt.show()

plt.figure()
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare (Colored by Survival)")
plt.show()
```
